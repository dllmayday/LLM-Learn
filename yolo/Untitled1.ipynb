## 使用自定义数据进行训练
### 1. 创建数据集
YOLOv5 模型必须使用已标注的数据进行训练，才能学习数据中的对象类别。在开始训练之前，有两种方法可以创建数据集：

使用Roboflow创建 YOLO 格式的数据集
#### 1.1 收集图像
你的模型将通过实例学习。使用与实际应用中类似的图像进行训练至关重要。理想情况下，你应该收集各种在相同配置（相机、角度、光照等）下拍摄的图像，这些配置应与你最终部署项目的配置相同。

如果无法做到这一点，您可以[从公共数据集](https://universe.roboflow.com/?ref=ultralytics)开始训练您的初始模型，然后在推理过程中从实际环境中采样图像，以迭代地改进您的数据集和模型。

本次演示下载一个[声纳图像数据集](https://universe.roboflow.com/chaitanya-8dheb/sonar-final/dataset/1)用于训练.

### 2. 选择一个预训练模型YOLOv5s作为训练起点

```
# Train YOLOv5s on COCO128 for 3 epochs
$ python train.py --img 640 --batch 16 --epochs 3 --data coco128.yaml --weights yolov5s.pt
```

该命令用于训练 YOLOv5 模型，各参数含义如下：

| 参数 | 含义 | 说明 |
| :--- | :--- | :--- |
| `--img 640` | 输入图像尺寸 | 将训练图像缩放至 640×640 像素。YOLOv5 默认采用正方形输入，该值通常为 32 的倍数（如 320、416、512、640 等），影响检测精度与显存占用。 |
| `--batch 16` | 批次大小 | 每个训练批次同时处理 16 张图像。较大的 batch 可加速训练，但需更多显存。 |
| `--epochs 3` | 训练轮数 | 整个数据集将被完整训练 3 次。这里设为 3 通常用于快速测试（验证流程），实际训练可能需要 100~300 轮才能收敛。 |
| `--data coco128.yaml` | 数据集配置文件 | 指向一个 YAML 文件，其中定义了数据集路径、类别数、类别名称等信息。`coco128.yaml` 是 YOLOv5 自带的一个小型演示数据集（128 张 COCO 子集）。 |
| `--weights yolov5s.pt` | 预训练权重文件 | 加载 `yolov5s.pt` 作为初始权重。`yolov5s` 是 YOLOv5 的小型模型（s=small），提供较好的速度与精度平衡。使用预训练权重可加速收敛、提升最终效果（迁移学习）。 |

此外，还有一些常用可选参数（未在该命令中出现但值得了解）：
- `--device`：指定训练设备，如 `0`（GPU 0）或 `cpu`。
- `--workers`：数据加载的子进程数。
- `--cache`：是否将数据集缓存至内存以加速训练。
- `--project` / `--name`：指定结果保存路径与实验名称。

该命令主要用于快速验证训练流程是否正常。实际训练时，通常会增加轮数（如 `--epochs 300`）并调整 `--batch` 以适应当前 GPU 显存。
