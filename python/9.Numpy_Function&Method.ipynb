## Functions and methods overview 函数和方法概述
[functions and methods names ordered in categories](https://numpy.org/doc/stable/reference/routines.html#routines)
   这是一个非常全面的 NumPy 函数分类整理。以下是按你提供的类别整理的函数列表及简要说明：

## 📦 Array Creation（数组创建）
| 函数 | 说明 |
|------|------|
| `arange` | 类似 range，返回数组 |
| `array` | 从列表/元组创建数组 |
| `copy` | 复制数组 |
| `empty` | 创建未初始化的数组（内容随机） |
| `empty_like` | 根据另一个数组形状创建空数组 |
| `eye` | 创建单位矩阵 |
| `fromfile` | 从二进制或文本文件读取数据 |
| `fromfunction` | 根据函数和形状创建数组 |
| `identity` | 创建方阵单位矩阵（类似 eye） |
| `linspace` | 等间距生成指定数量的数值 |
| `logspace` | 对数刻度等间距生成数值 |
| `mgrid` | 密集多维网格（用于 N 维索引） |
| `ogrid` | 开放多维网格（节省内存） |
| `ones` | 全 1 数组 |
| `ones_like` | 根据另一个数组形状创建全 1 数组 |
| `r_` | 按行（第一维）串联切片对象 |
| `zeros` | 全 0 数组 |
| `zeros_like` | 根据另一个数组形状创建全 0 数组 |

## 🔄 Conversions（数据类型与形状转换）
| 函数 | 说明 |
|------|------|
| `ndarray.astype` | 转换数组数据类型（方法） |
| `atleast_1d` | 将输入转换为至少 1 维 |
| `atleast_2d` | 将输入转换为至少 2 维 |
| `atleast_3d` | 将输入转换为至少 3 维 |
| `mat` | 将输入解释为矩阵（类似 matrix，已不推荐） |

## ✂️ Manipulations（数组操作与重塑）
| 函数 | 说明 |
|------|------|
| `array_split` | 分割数组（允许不等分） |
| `column_stack` | 水平堆叠一维数组为列 |
| `concatenate` | 沿现有轴拼接数组 |
| `diagonal` | 返回对角线元素 |
| `dsplit` | 沿第三维分割 |
| `dstack` | 沿第三维堆叠 |
| `hsplit` | 水平（列）方向分割 |
| `hstack` | 水平（列）方向堆叠 |
| `ndarray.item` | 将数组元素转换为 Python 标量 |
| `newaxis` | 在索引中添加新维度 |
| `ravel` | 展平数组（可能返回视图） |
| `repeat` | 重复数组元素 |
| `reshape` | 改变形状 |
| `resize` | 改变形状（可改变大小，会重复数据） |
| `squeeze` | 移除长度为 1 的轴 |
| `swapaxes` | 交换两个轴 |
| `take` | 从沿轴的位置取元素 |
| `transpose` | 转置（轴对换） |
| `vsplit` | 垂直（行）方向分割 |
| `vstack` | 垂直（行）方向堆叠 |

## ❓ Questions（条件判断与搜索）
| 函数 | 说明 |
|------|------|
| `all` | 是否所有元素为真 |
| `any` | 是否存在任意元素为真 |
| `nonzero` | 返回非零元素的索引 |
| `where` | 根据条件返回索引或选择元素 |

## 🔢 Ordering（排序与极值）
| 函数 | 说明 |
|------|------|
| `argmax` | 返回最大值的索引 |
| `argmin` | 返回最小值的索引 |
| `argsort` | 返回排序后的索引 |
| `max` | 最大值 |
| `min` | 最小值 |
| `ptp` | 极差（最大值-最小值） |
| `searchsorted` | 查找元素应插入的索引以保持顺序 |
| `sort` | 排序 |

## ⚙️ Operations（元素级与数组运算）
| 函数 | 说明 |
|------|------|
| `choose` | 从多个数组中选择元素构造新数组 |
| `compress` | 按条件选择元素 |
| `cumprod` | 累积乘积 |
| `cumsum` | 累积和 |
| `inner` | 内积 |
| `ndarray.fill` | 用标量填充数组（方法） |
| `imag` | 返回复数数组的虚部 |
| `prod` | 计算乘积 |
| `put` | 将指定值放入数组的指定位置 |
| `putmask` | 根据掩码将指定值放入数组 |
| `real` | 返回复数数组的实部 |
| `sum` | 求和 |

## 📐 Basic Linear Algebra（基础线性代数）
| 函数 | 说明 |
|------|------|
| `cross` | 向量叉积 |
| `dot` | 点积（矩阵乘法，1D/2D） |
